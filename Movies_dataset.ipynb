```python
import pandas as pd
```

```python
movies=pd.read_csv("movies.csv")
movies.shape
```

```python
movies.head()
```

```python
movies=movies.drop(["Released_Year","Certificate","Runtime","No_of_Votes","Gross","Meta_score"],axis=1)
```

```python
movies.isnull().sum()
```

```python
movies.head()
```

```python
movies.duplicated().sum()
```

```python
def convert(obj):
    l=[]
    for i in obj.split(','):
        l.append(i)
    return l
```

```python
movies['Genre']=movies['Genre'].apply(convert)
```

```python
def convert_o(obj):
    l=[]
    for i in obj.split(" "):
        l.append(i)
    return l
```

```python
movies['Director'] = movies['Director'].apply(convert)
movies['Star1'] = movies['Star1'].apply(convert)
movies['Star2'] = movies['Star2'].apply(convert)
movies['Star3'] = movies['Star3'].apply(convert)
movies['Star4'] = movies['Star4'].apply(convert)
```

```python
movies.head()
```

```python
def convert_r(obj):
    l=[]
    if obj:
        l.append(obj)
    return l
```

```python
movies['IMDB_Rating'] = movies['IMDB_Rating'].apply(convert_r)
```

```python
movies['Overview'] = movies['Overview'].apply(convert_o)
```

```python
movies.head()
```

```python
movies['Genre'] = movies['Genre'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['Director'] = movies['Director'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Star1'] = movies['Star1'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Star2'] = movies['Star2'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Star3'] = movies['Star3'].apply(lambda x:[i.replace(" ","") for i in x])
movies['Star4'] = movies['Star4'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
columns = ["Genre","IMDB_Rating","Star1","Star2","Star3","Star4","Overview","Director"]
movies['Tags'] = movies[columns[0]] + movies[columns[2]] + movies[columns[3]] + movies[columns[4]] + movies[columns[5]] + movies[columns[6]] + movies[columns[7]]
```

```python
movies['Tags'][0]
```

```python
new_df=movies[["Poster_Link","Series_Title","Tags"]]
```

```python
new_df.head()
```

```python
def cvrt(obj):
    l=""
    for i in obj:
        l+=" "
        if isinstance(i, float):
            l += str(i)
        else:
            l+=i
    return l
new_df['Tags'] = new_df['Tags'].apply(cvrt)
```

```python
new_df['Tags'][0]
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def st(obj):
    y=[]
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tags'] = new_df['Tags'].apply(st)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words='english')
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#for cosine distance
similarity = cosine_similarity(vectors)
#for euclidean distance
import numpy as np
euclidean_dist = np.sqrt(np.clip(2 * (1 - similarity), 0, None))
```

```python
euclidean_dist
```

```python
new_df[new_df['Series_Title'] == 'Giant'].index[0]
```

```python
def recommendation(entity):
    entity_index=new_df[new_df['Series_Title'] == entity].index[0]
    distances=similarity[entity_index]
    entity_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:10]
    for i in entity_list:
        print(new_df.iloc[i[0]].Series_Title)
```

```python
new_df['Series_Title'].tolist()
```

```python
recommendation('Iron Man')
```
