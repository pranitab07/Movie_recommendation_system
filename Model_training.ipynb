```python
import pandas as pd
```

```python
df = pd.read_csv("bollywood_full.csv")
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(['tagline','wins_nominations','release_date','is_adult','year_of_release','runtime','title_y','original_title'],axis=1)
```

```python
df.head()
```

```python
df = df.drop('imdb_votes',axis=1)
df = df.dropna()
```

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.duplicated().sum()
df.shape
```

```python
def convert(obj):
    l=[]
    for i in obj.split('|'):
        l.append(i)
    return l
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
def convert_a(obj):
    l=[]
    for i in obj.split('|'):
        l.append(i)
    return l
```

```python
df['actors'] = df['actors'].apply(convert_a)
```

```python
def cvrt(obj):
    l=[]
    if isinstance(obj,float):
        l.append(obj)
    return l
```

```python
df['imdb_rating'] = df['imdb_rating'].apply(cvrt)
```

```python
df.head()
```

```python
def convert_s(obj):
    l=[]
    for i in obj.split(" "):
        l.append(i)
    return l
```

```python
df['story'] = df['story'].apply(convert_s)
df['summary'] = df['summary'].apply(convert_s)
```

```python
df.head()
```

```python
df['actors'] = df['actors'].apply(lambda x: [i.replace(" ", "") for i in x])
df['genres'] = df['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
df.head()
```

```python
columns=['genres','imdb_rating','story','summary','actors']
df['Tags'] = df[columns[0]]+df[columns[1]]+df[columns[2]]+df[columns[3]]+df[columns[4]]
```

```python
new_df = df[['poster_path','wiki_link','title_x','imdb_id','Tags']]
new_df.head()
```

```python
def cvrt(obj):
    l=""
    for i in obj:
        l+=" "
        if isinstance(i, float):
            l += str(i)
        else:
            l+=i
    return l 
new_df['Tags'] = new_df['Tags'].apply(cvrt)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
new_df['Tags'] = new_df['Tags'].apply(lambda x:x.lower())
new_df['Tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
Ps = PorterStemmer()
```

```python
def stem(obj):
    y=[]
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tags'].apply(stem)[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000,stop_words='english')
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#for cosine distance
similarity = cosine_similarity(vectors)
#for euclidean distance
import numpy as np
euclidean_dist = np.sqrt(np.clip(2 * (1 - similarity), 0, None))
```

```python
new_df[new_df['title_x'] == 'Uri: The Surgical Strike'].index[0]
```

```python
def recommendation(entity):
    entity_index=new_df[new_df['title_x'] == entity].index[0]
    distances=similarity[entity_index]
    entity_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[0:5]
    for i in entity_list:
        print(new_df.iloc[i[0]].title_x)
```

```python
recommendation('The Accidental Prime Minister (film)')
```
